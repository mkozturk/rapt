# Adaptive switching between particle and guiding-center modes - Speiser orbits

We have an `Adaptive` class built on top of `Particle` and `GuidingCenter` classes. It switches to the appropriate method by checking the local adiabatic conditions. Internally stores the data as a list of `Particle` and `GuidingCenter` objects. The interface is identical to the other classes.

```python
%matplotlib inline
import numpy as np
import pylab as pl
from mpl_toolkits.mplot3d import Axes3D
import rapt
```

```python
pl.rcParams['figure.figsize'] = (10.0, 8.0)
```

Consider a parabolic tail-like magnetic field:

$\vec{B} = B_0 \frac{z}{d} \hat{x} + B_n \hat{z}$ if $|z| \le 1$

$\vec{B} =  \mathrm{sgn}(z) B_0 \hat{x} + B_n \hat{z}$ if $|z| \ge 1$

with $B_0 = 10$, $B_n = 1$, $d=0.2$.

Examine Speiser orbits under this field.

```python
f = rapt.fields.Parabolic()
rapt.params['solvertolerances'] = (1e-12,1e-12)  # need to be accurate for this problem
rapt.params['epss'] = 0.02   # spatial adiabaticity parameter
rapt.params['Ptimestep'] = 0.1   # time step to sample the particle motion (seconds)
rapt.params["GCtimestep"] = 1    # time step to sample the guiding-center motion (seconds)
```

Initialize a Particle and an Adaptive with the same conditions, trace and compare.

```python
p = rapt.Particle([5, -5, 0.9], [-0.1,0.1,0], 0, mass=1, charge=1, field=f)
pa = rapt.Adaptive([5, -5, 0.9], [-0.1,0.1,0], 0, mass=1, charge=1, field=f)
```

```python
pa.advance(300)
```

Now trace the particle.

```python
p.advance(300)
```

Display the trajectories

```python
fig = pl.figure()
ax = Axes3D(fig)
ax.plot( p.getx(), p.gety(), p.getz(), label='Particle')
ax.plot( pa.getx(), pa.gety(), pa.getz(), 'r', label='Adaptive')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Parabolic tail field')
ax.legend()
```

(To view the trajectory in an interactive window give the IPython command `%matplotlib qt`, and rerun the previous cell.)

We see that the Particle agrees well with the Adaptive trajectory up until the current sheet. Then, the Adaptive mode sets up a Particle with an arbitrary phase angle. Because of the chaos inherent in the field geometry, the two orbits diverge.

Plot the kinetic energy in the Adaptive mode:

```python
%matplotlib inline
```

```python
pl.plot(pa.gett(), pa.getke())
pl.title("Kinetic energy vs time")
pl.xlabel("time [s]")
pl.ylabel("KE [J]");
```

It can be seen that the trajectory loses more precision during the Particle mode compared to the GuidingCenter mode.
