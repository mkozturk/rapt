# Bounce-averaged drift motion: Drift-shell splitting

```python
%matplotlib inline
import numpy as np
import pylab as pl
pl.rcParams['figure.figsize'] = (12.0, 9.0)
pl.style.use("ggplot")
```

```python
import time
import rapt
from rapt import m_pr, e, re, Re
```

Suppose we initialize two identical particles that have identical initial position and speed, but differing in pitch angle. If the field is not symmetrical with respect to the azimuthal angle, these two particles will follow different drift shells. This phenomenon is called _drift-shell splitting_.

This notebook illustrates drift-shell splitting using the double-dipole model.

```python
f = rapt.fields.DoubleDipole()
```

```python
rapt.params['solvertolerances'] = (1e-5, 1e-5)
rapt.params['BCtimestep'] = 0.5  # in units of the initial bounce period

pos = (-7.8*Re,0,0)
v = rapt.utils.speedfromKE(1e5, m_pr) # speed of an 100keV proton

b1 = rapt.BounceCenter(pos,v, pa=90, mass=m_pr, charge=e, field=f)
b2 = rapt.BounceCenter(pos,v, pa=10, mass=m_pr, charge=e, field=f)
```

```python
t1 = time.time()
b1.advance(4000)
t2 = time.time()
print('computing time: ', t2-t1)
```

```python
t1 = time.time()
b2.advance(2000)
t2 = time.time()
print('computing time: ', t2-t1)
```

```python
%matplotlib qt
```

```python
pl.plot(b1.getx()/Re, b1.gety()/Re, label="pa=90")
#pl.plot(b2.getx()/Re, b2.gety()/Re, label="pa=10")
pl.title('Bounce-averaged drift paths under double dipole')
pl.xlabel('x [Re]')
pl.ylabel('y [Re]')
pl.legend()
pl.grid()
```

Let us repeat the same calculation with using `GuidingCenter` tracer.

```python
rapt.params['solvertolerances'] = (1e-6, 1e-6)
rapt.params["bounceresolution"] = 20
g1 = rapt.GuidingCenter(pos,v, pa=90, mass=m_pr, charge=e, field=f)
g2 = rapt.GuidingCenter(pos,v, pa=10, mass=m_pr, charge=e, field=f)
```

```python
t1 = time.time()
g1.advance(4500)
t2 = time.time()
print('computing time: ', t2-t1)
```

```python
def tmp(self):
    """Return a 1-d array of total magnetic field strength values along the 
    trajectory."""
    out = np.zeros(len(self.trajectory))
    for i, row in enumerate(self.trajectory):
        out[i] = self.field.magB(row[:4])
    return out
```
