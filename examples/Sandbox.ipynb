```python
%matplotlib inline
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d import Axes3D

pl.style.use("ggplot")

import rapt
```

```python
# Initialize the field
f = rapt.fields.EarthDipole()

# Initialize a particle.
# All values in SI units.
p = rapt.Particle(
    pos=(6*rapt.Re, 0,0),  # x, y, z
    vel=(0, -1e7, 2e7),    # vx, vy, vz
    t0=0,                  # initial time
    mass=rapt.m_pr,        # m_pr : proton mass, m_el : electron mass
    charge=rapt.e,         # e: elementary charge
    field=f)
```

```python
rapt.params['cyclotronresolution'] = 20  # points per cyclotron period
```

```python
p.advance(100)
```

```python
pl.plot(p.getx()/rapt.Re, p.gety()/rapt.Re)
pl.xlabel("x [Re]")
pl.ylabel("y [Re]")
pl.title("Proton at 6 Re");
```

```python
pl.plot(p.gett(), p.getz()/rapt.Re)
pl.xlabel("time [sec]")
pl.ylabel("z [Re]")
pl.title("Proton at 6 Re");
```

```python
fig = pl.figure()
ax = Axes3D(fig)
ax.plot( p.getx()/rapt.Re, p.gety()/rapt.Re, p.getz()/rapt.Re, label='Particle')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Proton around Earth');
```
