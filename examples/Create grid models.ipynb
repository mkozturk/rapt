# How to add a new grid-based field model

RAPT supports two types of field models: Analytic and grid-based.

Analytic models express $\vec{E}(t,x,y,z)$ and $\vec{B}(t,x,y,z)$ as explicit functions of time and position. However, many magnetospheric models derive the fields from complicated calculations, such as MHD simulations or empirical regressions of data. Such models can be assimilated to RAPT by sampling them on a grid.

This notebook shows how to create a new grid field class. We will illustrate it using the RCME field model.

The data for the grid is stored on one or more files on disk. Each file contains the grid for a specific time point. Grids must be rectangular, not necessarily with uniform spacing, and the grid structure must stay the same at all time points. The input to the grid field is a list of such files, in order of time.

How the data is organized is up to the user. The field will call a method `parsefile` which reads the data file and returns the relevant data organized in a data dictionary. The details of the data file should be handled in this method.

To preserve memory, the field is interpolated using three consecutive grids. The 4D interpolation is updated as time proceeds, and earlier time points are discarded.

The interpolated field values are accessed with the `Egrid` and `Bgrid` methods. By default, `E` and `B` are wrappers around these, but in practice they contain code to add a dipole component or to handle symmetries.

In the RCME field below, the grid data includes the dipole field component as well as other external fields. During parsing in `parsefile`, the dipole component is subtracted, and the interpolation is done with external fields only. The external fields vary much more smoothly, so the errors arising from the linear interpolation are reduced. At the front end, the `B` function adds this dipole component to the interpolated field given by `Bgrid`. This trick is recommended for all other fields.

```python
import re   # regular expressions library - for parsing data
import numpy as np
from rapt import Re
from rapt.fields import Grid
```

```python
class RCME(Grid):
    B0 = 30574*1e-9 # the mean value of the magnetic field at the magnetic equator on the Earth's surface (Tesla)
    _coeff = -3*B0*Re**3
    
    def __init__(self,filelist):
        Grid.__init__(self,filelist)
    
    def parsefile(self,fname):   # reads a single data file (one time point)
        f = open(fname)
        f.readline()
        f.readline()
        thirdline = f.readline()
        # Read the metadata
        m = re.search(r'FRICTION-IT=(\d+).*I=(\d+).*J=(\d+).*K=(\d+)',thirdline)
        time = float(m.group(1))*60
        nx = int(m.group(2))
        ny = int(m.group(3))
        nz = int(m.group(4))
        
        # Data arrays to fill up:
        x = np.zeros(nx)
        y = np.zeros(ny)
        z = np.zeros(nz)
        Bx = np.zeros((nx,ny,nz))
        By = np.zeros((nx,ny,nz))
        Bz = np.zeros((nx,ny,nz))
        Ex = np.zeros((nx,ny,nz))
        Ey = np.zeros((nx,ny,nz))
        Ez = np.zeros((nx,ny,nz))
        
        # Read each line and fill up arrays using indices I,J,K.
        for line in f:
            L = [float(a) for a in line.split()]
            i,j,k,xi,yj,zk,inside,bx,by,bz,ex,ey,ez = L
            # Subtract 1 from the index values to make them 0-base.
            i = int(i-1)
            j = int(j-1)
            k = int(k-1)
            x[i] = xi
            y[j] = yj
            z[k] = zk
            Bx[i,j,k] = bx
            By[i,j,k] = by
            Bz[i,j,k] = bz
            Ex[i,j,k] = ex
            Ey[i,j,k] = ey
            Ez[i,j,k] = ez
        f.close()
        
        # The first and last element in position arrays should be discarded.
        x = x[1:-1]
        y = y[1:-1]
        z = z[1:-1]
        Bx = Bx[1:-1, 1:-1, 1:-1]
        By = By[1:-1, 1:-1, 1:-1]
        Bz = Bz[1:-1, 1:-1, 1:-1]
        Ex = Ex[1:-1, 1:-1, 1:-1]
        Ey = Ey[1:-1, 1:-1, 1:-1]
        Ez = Ez[1:-1, 1:-1, 1:-1]
        
        # To reduce interpolation errors, 
        # remove the dipole component, output only the residual.        
        X, Y, Z = np.meshgrid(x, y, z, indexing='ij')
        R2 = X**2 + Y**2 + Z**2
        R5inv = R2 ** (-5/2)
        rr = 1
        coeff = (-3*self.B0*rr**3)*R5inv
        Bdipx = coeff * X*Z
        Bdipy = coeff * Y*Z
        Bdipz = coeff * (Z**2 - R2/3)
        
        # The output must be a dictionary with all the keys below, except 'B0'.
        return dict({'time':time,
                     'B0': self.B0, # model-specific, not required for interpolation
                     'x': x*Re,
                     'y': y*Re,
                     'z': z*Re,
                     'Bx': Bx - Bdipx, 
                     'By': By - Bdipy,
                     'Bz': Bz - Bdipz,
                     "Ex": Ex*1e-3, # convert from mV/m to V/m
                     "Ey": Ey*1e-3, 
                     "Ez": Ez*1e-3})
    
    def B(self, tpos):
        t,x,y,z = tpos[:]
        Bnorth = self.Bgrid([t,x,y,abs(z)])  # grid defined only for z>=0
        # Use the symmetry to handle negative z values:
        if z<0:
            Bg = Bnorth * [-1,-1,1]   # Bx -> -Bx, By -> -By, Bz -> Bz
        else:
            Bg = Bnorth
        # Add the dipole component
        r2 = x*x+y*y+z*z
        Bdip =  self._coeff / pow(r2, 2.5) * np.array([x*z, y*z, (z*z-r2/3)])
        return Bg+Bdip
    
    def E(self,tpos):
        t,x,y,z = tpos[:]
        Enorth = self.Egrid([t,x,y,abs(z)]) # grid defined only for z>=0
        # Use the symmetry to handle negative z values:
        if z<0:
            return Enorth * [1,1,-1] # Ex -> -Ex, Ey -> -Ey, Ez -> Ez
        else:
            return Enorth
```
