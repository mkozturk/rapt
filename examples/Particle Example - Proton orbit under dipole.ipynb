# Proton orbits under dipole magnetic field

```python
%matplotlib inline
import pylab as pl
pl.style.use("ggplot")
```

```python
import rapt
```

Initialize the field object

```python
f = rapt.fields.EarthDipole()
```

## Equatorial proton

A Particle object is used to trace a particle in a given magnetic field. Initialize it with an equatorial proton at $6R_e$ under a dipole field.

```python
p = rapt.Particle(pos=(6*rapt.Re, 0,0), 
                  vel=(0, -1e7, 0), 
                  t0=0, 
                  mass=rapt.m_pr, 
                  charge=rapt.e, 
                  field=f)
```

The `advance()` method of the `Particle` object solves the Larmor equation.

```python
p.advance(1)   # Advance the solution for 1 second.
```

To plot the trajectory in x-y coordinates, use the `getx()` and `gety()` methods.

```python
pl.plot(p.getx()/rapt.Re, p.gety()/rapt.Re)
pl.xlabel("x [Re]")
pl.ylabel("y [Re]")
pl.title("Proton at 6 Re");
```

The complete solution is stored internally in the `p.trajectory` variable. If the `advance()` method is called again, it solves the equation starting from the final point in `p.trajectory`, and extends it.

```python
p.advance(1)
```

```python
pl.plot(p.getx()/rapt.Re, p.gety()/rapt.Re)
pl.xlabel("x [Re]")
pl.ylabel("y [Re]")
pl.title("Proton at 6 Re");
```

The resolution can be changed using the `params` dictionary.

```python
rapt.params['cyclotronresolution'] = 20  # points per cyclotron period
```

Follow the particle for 2 more seconds with finer sampling.

```python
p.advance(2)
```

```python
pl.plot(p.getx()/rapt.Re, p.gety()/rapt.Re)
pl.xlabel("x [Re]")
pl.ylabel("y [Re]")
pl.title("Proton at 6 Re");
```

Plot the kinetic energy of the proton vs. time, as a measure of the precision of the solution.

```python
pl.plot(p.gett(), p.getke())
pl.xlabel("time [s]")
pl.ylabel("K [J]");
```

## Specifying the pitch angle and kinetic energy

Usually we specify a particle's initial conditions not with the velocity vector, but with the kinetic energy and the pitch angle (angle between the magnetic field vector and the velocity).

```python
from numpy import pi,sin, cos
ke = 1e6  # 1MeV electron-volts kinetic energy
pa = 60   # degrees
v = rapt.utils.speedfromKE(ke, rapt.m_pr)  # speed corresponding to the given relativistic kinetic energy
vel = [v*sin(pa*pi/180), 0, v*cos(pa*pi/180)]
```

```python
p = rapt.Particle(pos=(6*rapt.Re, 0,0), 
                  vel=vel, 
                  t0=0, 
                  mass=rapt.m_pr, 
                  charge=rapt.e, 
                  field=f)
```

There is an easier way by using the `setke` and `setpa` methods.

```python
p = rapt.Particle(pos=(6*rapt.Re, 0,0), 
                  vel=[1,0,1],    # just make up a provisional velocity
                  t0=0, 
                  mass=rapt.m_pr, 
                  charge=rapt.e, 
                  field=f)
p.setpa(60)
p.setke(1e6);
```

```python
p.advance(100)
```

Plot the orbit in rectangular projection using the methods `gettheta` (longitude) and `getphi` (colatitude).

```python
pl.plot(p.gettheta()/pi*180, 90-p.getphi()/pi*180)
pl.xlabel("longitude")
pl.ylabel("latitude")
pl.title("Proton at 6 Re");
```

Check the conservation of energy.

```python
pl.plot(p.gett(), p.getke()/rapt.e/1e6)
pl.xlabel("time [s]")
pl.ylabel("K [MeV]");
```

Check the conservation of the magnetic moment (first invariant), as calculated at the guiding-center of the motion.

```python
pl.plot(p.gett(), p.mu())
pl.xlabel("time [s]")
pl.ylabel("$\mu$ [J/T]");
```

Due to the inherent truncation errors in determining the guiding center, the magnetic moment oscillates with a small amplitude.

## Saving and loading Particle instances

You may want to save a `Particle` object on disk for later processing. The `save` method of `Particle` can be used for this purpose.

```python
p.save("pr_1mev_pa60_L6")
```

The Particle instance and all the data contained in it is now stored in the file `pr_1mev_pa60_L6` in the current directory. To load it back, we create a new instance of the `Particle` class and use the `load` method.

```python
p2 = rapt.Particle()
p2.load("pr_1mev_pa60_L6")
```

Plot the trajectory of the loaded particle to verify.

```python
pl.plot(p2.gettheta()/pi*180, 90-p2.getphi()/pi*180)
pl.xlabel("longitude")
pl.ylabel("latitude")
pl.title("Proton at 6 Re");
```

The `save` and `load` methods use the built-in `pickle` module in a straightforward way. An alternative way of loading a `Particle` from the disk is as follows.

```python
import pickle
f = open("pr_1mev_pa60_L6","rb")
p3 = pickle.load(f)
f.close()

pl.plot(p3.gettheta()/pi*180, 90-p3.getphi()/pi*180)
pl.xlabel("longitude")
pl.ylabel("latitude")
pl.title("Proton at 6 Re");
```
