# Plotting field lines of the double-dipole model

```python
%matplotlib inline
import numpy as np
import pylab as pl
pl.style.use("ggplot")
```

```python
import rapt
```

Initialize the double dipole field.

```python
f = rapt.fields.DoubleDipole()
```

Generate a set of field lines starting at various locations on the x-axis.

```python
fls = []
for x in np.arange(5,10,0.5):  # x = 5, 5.5, 6, ..., 9.5
    fl = rapt.fieldline.Fieldline((0, x*rapt.Re, 0, 0), f, Bmax = rapt.B0)
    fl.trace()
    fls.append(fl)
for x in np.arange(-9,-5,0.5): # x = -9, -8.5, ..., -5.5
    fl = rapt.fieldline.Fieldline((0, x*rapt.Re, 0, 0), f, Bmax = rapt.B0)
    fl.trace()
    fls.append(fl)
```

Plot the field lines on the x-z plane.

```python
for fl in fls:
    pl.plot(fl.getx()/rapt.re, fl.getz()/rapt.re,"g")
    pl.xlabel("x [Re]")
    pl.ylabel("z [Re]")
```

Trace a field line close to the magnetopause, and plot the field strength along the field line. We see that the compression of lines causes a local maximum of field strength at the equator (s=0).

```python
x = 9.5
fl = rapt.fieldline.Fieldline((0, x*rapt.Re, 0, 0), f, Bmax = rapt.B0)
fl.trace()
pl.plot(fl.gets()/rapt.re, fl.getB()*1e9)  # Plot B vs s.
pl.xlabel("Distance along field line [Re]")
pl.ylabel("Field strength [nT]")
pl.ylim([0,100]);
```

We see that the field strength $B(s)$ has a local maximum at $s=0$ and two local minima at the sides. The utility function `flutils.criticalpoints` returns the location, field strength, and the second derivative of field strength for each of these points.

```python
rapt.flutils.criticalpoints((0, 9.5*rapt.Re, 0, 0), f, Bm = rapt.B0)
```
